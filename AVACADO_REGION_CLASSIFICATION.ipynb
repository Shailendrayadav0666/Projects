```python
df.drop('Date',inplace=True,axis=1)
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['type']=le.fit_transform(df['type'])
df['region']=le.fit_transform(df['region'])
df
```

```python
for col in df.columns:
    if df.skew().loc[col]>0.55:
        df[col]=np.log1p(df[col])
```

```python
df.skew()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
x=df_new.iloc[:,:-1]
print(x)
```

```python
y=df_new.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
model=MultinomialNB(),SVC(),DecisionTreeClassifier()
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

```python
from sklearn.model_selection import cross_val_score
dt=DecisionTreeClassifier()
dtc=dt.fit(x_train,y_train)
print(dtc.score(x_train,y_train))

print('Mean cross_val_score is=',cross_val_score(dt,x,y,cv=10).mean())
print('STD cross_val_score is=',cross_val_score(dt,x,y,cv=10).std())
```


# Regression-with label as avg price

```python
x=df_new.iloc[:,2:]
x
```

```python
y=df_new.iloc[:,1]
y
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2 score corresponding to ',r_state,'is',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
        
    print('max r2_score corresponding to',final_r_state,'is',max_r_score)
    return final_r_state
```

```python
lreg=LinearRegression()
maxr2_score(lreg,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
neighbors={'n_neighbors':range(1,30)}
knr=KNeighborsRegressor()
gknr=GridSearchCV(knr,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=13)
r_state=maxr2_score(knr,x,y)
```

```python
from sklearn.linear_model import Lasso
lsreg=Lasso()
parameters={'alpha':[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
lsreg=Lasso(alpha=0.001)
r_state=maxr2_score(lsreg,x,y)
```

```python
from sklearn.model_selection import cross_val_score


print ('mean of KNN regressor is',cross_val_score(knr,x,y,cv=5,scoring='r2').mean())
print ('std deviation of the knn regressor is ',cross_val_score(knr,x,y,cv=5,scoring='r2').std())
```
