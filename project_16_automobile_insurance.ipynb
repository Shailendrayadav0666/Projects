```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2 score corresponding to ',r_state,'is',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
        
    print('max r2_score corresponding to',final_r_state,'is',max_r_score)
    return final_r_state
    
```

```python
lr=LinearRegression()
maxr2_score(lr,x,y)
```

```python
cvsl=cross_val_score(lr,x,y,cv=10,scoring='r2').mean()
cvsl
```

```python
knr=KNeighborsRegressor()
n_neighbors={'n_neighbors':range(1,30)}
gknr=GridSearchCV(knr,n_neighbors,cv=10)
gknr.fit(x,y)
print(gknr.score(x,y))
print(gknr.best_params_)
```

```python
knr=KNeighborsRegressor(n_neighbors=5)
maxr2_score(knr,x,y)
```

```python
knrc=cross_val_score(knr,x,y,cv=10,scoring='r2').mean()
knrc
```

```python
dt=DecisionTreeRegressor()
maxr2_score(dt,x,y)
```

```python
dtc=cross_val_score(dt,x,y,cv=10,scoring='r2')
dtc.mean()
```

```python
dtc.std()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
maxr2_score(rf,x,y)
```

```python
rfc=cross_val_score(rf,x,y,cv=10,scoring='r2')
rfc.mean()
```
